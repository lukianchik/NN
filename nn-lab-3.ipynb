# **Предобработка**

Убираем все строки, где есть пропущенные значения

```python
training_data = training_data.dropna()
test_data = test_data.dropna()
```

```python
training_data.head()
```

```python
test_data.info()
```

Кодируем One hot encoder из pandas столбцы с текстовыми значениями

```python
text_type = ['title_status', 'transmission', 'drive', 'size']
training_data = pd.get_dummies(training_data, prefix=text_type)
test_data = pd.get_dummies(test_data, prefix=text_type)
```

```python
training_data.head()
```

Разделяем данные на фичи и цели, также я добавил некоторое разделение с train_test_split, чтобы в конце предсказать точность модели

```python
import numpy as np
from sklearn.model_selection import train_test_split

target_variable_name = 'price'
X = training_data.drop(target_variable_name, axis=1).astype(int)
y = training_data[target_variable_name].astype(int)

X_test = test_data.drop(target_variable_name, axis=1).astype(int)
y_test = test_data[target_variable_name].astype(int)

X_train, X_pred, y_train, y_pred = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Задание (повышенный уровень):**
Создать нейронную сеть, которая будет предсказывать цену автомобиля на основе всех полей исходного датасета (оценку качества регрессора проводить на основе метрик MAE, MSE). # **Создание нейронной сети**

Создаем нейронную сеть

```python
from tensorflow import keras
from tensorflow.keras import layers, callbacks

model = keras.Sequential([
    layers.Flatten(input_shape=[22]),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='linear')
])
```

Метрики по заданию MAE и MSE

```python
model.compile(
    optimizer='adam',
    loss='mean_absolute_error',
    metrics=["mean_absolute_error", "mean_squared_error"],
)
```

Тренируем нейросеть

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    batch_size=1024,
    epochs=45,
)
```

Посмотрим на точность модели

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print('MAE:', mean_absolute_error(y_pred, model.predict(X_pred)))
```
