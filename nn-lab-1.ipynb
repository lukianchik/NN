**Задание:**
Создать модель (без использования готовых решений) простейшего перцептрона, которая будет выполнять функцию логического "ИЛИ".

*Общее требование:*  Обучение и тестирование должно проходить на зашумленных данных (уровень шума +- 0.45). 

**Функции и генерация входных данных**

```python
import copy
import random
import numpy as np

#Функция активации
def result_bin_class(result):
    if (result > 1):
        return True
    else:
        return False
    
#Функция подсчета суммарного веса полученных сигналов
def calculate_res(input_number, first_layer_weights):
    sum = 0
    for i in range(2):
        sum += input_number[0][i] * first_layer_weights[i]
    return sum

#Функция, формирующая сигнал корректировки 
#весов в зависимости от соотношения ответа нейронной сети и правильного ответа
def weight_changing(positive, target):
    k = 0
    if (positive and not target):
        k = -0.1
    elif (not positive and target):
        k = 0.1
    return k

#Функция обучения нейронной сети
def learn(input_number, first_layer_weights, target):
    k = weight_changing(result_bin_class(calculate_res(input_number, first_layer_weights)), True if input_number[1] == target else False)
    for i in range(2):
        first_layer_weights[i] += input_number[0][i] * k
    return first_layer_weights

def train_nn(input_dataset, first_layer_weights, target, size_of_data):
    for i in range(size_of_data):
        first_layer_weights = learn(random.choice(input_dataset), first_layer_weights, target)
    
#Функция тестирования нейронной сети
def test(input_number, first_layer_weights, target):
    if (result_bin_class(calculate_res(input_number, first_layer_weights)) == (True if input_number[1] == target else False)):
        return 1
    else:
        return 0

def test_nn(input_dataset, first_layer_weights, target, size_of_data):
    res = 0
    for i in range(size_of_data):
        res += test(random.choice(input_dataset), first_layer_weights, target)
    return (res / size_of_data)

#Задание списка логического ИЛИ
dataset = list()
dataset.append(([0, 0],0))
dataset.append(([0, 1],1))
dataset.append(([1, 0],1))
dataset.append(([1, 1],1))

#Создание датасетов
noise_dataset = []
clear_dataset = []
for i in range(10000):
    noise_dataset.append(copy.deepcopy(random.choice(dataset)))
    clear_dataset.append(copy.deepcopy(random.choice(dataset)))
    for j in range(2):
        noise_dataset[i][0][j] += (random.randint(-45, 45) * 0.01)
```

**Обучение**

```python
noise_layer_weights = np.ones(2)
clear_layer_weights = np.ones(2)
train_nn(noise_dataset, noise_layer_weights, 1, 100000)
train_nn(clear_dataset, clear_layer_weights, 1, 100000)
```

**Подсчет точности**

```python
print('Accuracy-score на зашумленных данных: ', test_nn(noise_dataset, noise_layer_weights, 1, 100000))
print('Веса при зашумленных данных:', noise_layer_weights)
print('Accuracy-score на чистых данных: ', test_nn(clear_dataset, clear_layer_weights, 1, 100000))
print('Веса при чистых данных:', clear_layer_weights)
```

**Можно заметить, что веса при чистых данных корректируются минимально, чтобы проходить порог в функции result_bin_class и получать максимальную точность                          
При шумных данных веса корректируется сильнее, чтобы получить точность близкую к 1**
