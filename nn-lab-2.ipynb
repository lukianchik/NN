# **Импорт библиотек и генерация датасета**

```python
import random
import numpy as np

dataset = list()

actions = ['+','-']


for i in range(10000):
    temp = list()
    temp.append(random.randint(-100, 100) * 0.01)
    temp.append(random.randint(-100, 100) * 0.01)
    temp.append(random.randint(0, 1))
    if temp[2]:
        if temp[0] - temp[1] < 0:
            temp.append(0)
        elif temp[0] - temp[1] == 0:
            temp.append(1)
        else:
            temp.append(2)
    else:
        if temp[0] + temp[1] < 0:
            temp.append(0)
        elif temp[0] + temp[1] == 0:
            temp.append(1)
        else:
            temp.append(2)
        dataset.append(temp)
```

# **Написание модели**

```python
# Задание весов
first_layer_weight_multiclass = np.zeros((4, 3))
first_layer_weight_multiclass
```

```python
# Вычисление суммы взвешанных значений(входное умножить на вес)
def calculate_res_multiclass(input_number, first_layer_weights):
    sum_array = [0, 0, 0] 
    for j in range(3):
        for i in range(4):
            sum_array[j] += input_number[i] * first_layer_weights[i][j]
    return sum_array

# Выбор наибольшего значения(самого вероятного класса)
def result_multiclass(sum_array):
    return sum_array.index(max(sum_array))

# Обновление весов модели(обучение)
def learn_multiclass(input_number, first_layer_weights):
    if result_multiclass(calculate_res_multiclass(input_number, first_layer_weights)) != input_number[3]:
        for j in range(3):
            if j == input_number[3]:
                k = 0.5
            else:
                k = -0.1
            for i in range(4):
                first_layer_weights[i][j] += input_number[i] * k
    return first_layer_weights

# Обучение модели
def train_nn_mc(input_dataset, first_layer_weights, size_of_dataset):
    for _ in range(size_of_dataset):
        learn_multiclass(random.choice(input_dataset), first_layer_weights)

# Тестирование на одном наборе входных значений
def test_multiclass(input_number, first_layer_weights):
    if result_multiclass(calculate_res_multiclass(input_number, first_layer_weights)) == input_number[3]:
        return 1
    else:
        return 0
# Тестирование на всем наборе и вывод точности
def test_nn_mc(input_dataset, first_layer_weights, size_of_dataset):
    correct_predictions = 0
    for _ in range(size_of_dataset):
        correct_predictions += test_multiclass(random.choice(input_dataset), first_layer_weights)
    return correct_predictions / size_of_dataset

# Тренировка(обновление весов)
train_nn_mc(dataset, first_layer_weight_multiclass, 8000)
# Расчет точности
print('Accuracy-score: ', test_nn_mc(dataset, first_layer_weight_multiclass, 2000))
```
